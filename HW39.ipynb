```python
from bs4 import BeautifulSoup
import requests
import asyncio, aiohttp
import html2text
from urllib.parse import urljoin
```

```python
async def fetch_data():
  async with aiohttp.ClientSession() as session:
    for i in range(1, 51):
      url = f'http://books.toscrape.com/catalogue/page-{i}.html'
      async with session.get(url) as response:
        if response.status == 200:
          data = await response.text()
          text_content = html2text.html2text(data)
          with open(f'image_{i}.txt', 'w') as f:
                f.write(text_content)
                print(f"Сохранил текст с {i} ссылки сайта")
          soup = BeautifulSoup(data, 'html.parser')
          img_tags = soup.find_all('img')
          for j, img in enumerate(img_tags):
            img_url_relative = img['src']
            base_url = f'{url}'
            img_url_absolute = urljoin(base_url, img_url_relative)
            response = requests.get(img_url_absolute)
            if response.status_code == 200:
              with open(f'image_{i}_{j}.jpg', 'wb') as f:
                f.write(response.content)
                print(f"Сохранил {j} изображение с {i} ссылки")
```

```python
task1 = asyncio.create_task(fetch_data())
await task1
```
